```python
import os
import pandas as pd
import bibtexparser
from bibtexparser.bparser import BibTexParser
import yaml
```

```python
# assign directory
directory = 'G:\Meu Drive\MBA - Data Engeneering\Python for Data Engineer\Trabalho\Arquivos\dir'
 
df = []
for filename in os.scandir(directory):
    if filename.is_file():
        print(filename.path)
        with open(filename.path,encoding="utf8") as bibtex_file:
            bibtex_database = bibtexparser.load(bibtex_file)                
            df_temp = pd.DataFrame(bibtex_database.entries)
            df.append(df_temp)
            print (bibtex_database)
```

```python
df_final = pd.DataFrame()
for dataset in df:
    df_temp = pd.DataFrame(dataset, columns = ['author','title','keywords','abstract','year','ENTRYTYPE','doi'])
    df_temp.rename(columns = {'ENTRYTYPE':'type_publication'}, inplace = True)
    df_final = pd.concat([df_final,df_temp])
display(df_final)
```

```python
yml_config = """csv_dest: G:\Meu Drive\MBA - Data Engeneering\Python for Data Engineer\Trabalho\export_final.csv
json_dest: G:\Meu Drive\MBA - Data Engeneering\Python for Data Engineer\Trabalho\export_final.json
yaml_dest: G:\Meu Drive\MBA - Data Engeneering\Python for Data Engineer\Trabalho\export_final.yaml
"""

with open("G:\Meu Drive\MBA - Data Engeneering\Python for Data Engineer\Trabalho\Config\config_final.yaml", "w") as f:
    f.write(yml_config)
```

```python
with open("G:\Meu Drive\MBA - Data Engeneering\Python for Data Engineer\Trabalho\Config\config_final.yaml","r") as f:
    config = yaml.load(f, Loader=yaml.FullLoader)
    print("Read successful")
print(config)
```

```python
df_final.to_csv (config["csv_dest"], index = False, header=True)
```

```python
df_final.to_json(config["json_dest"],orient='records')
```

```python
export_yaml = yaml.dump(
    df_final.reset_index().to_dict(orient='records'),
    sort_keys=False, width=72, indent=4,
    default_flow_style=None)

with open(config["yaml_dest"], 'w') as f:
    f.write(export_yaml)
```
